```python
import pandas as pd
```

```python
df_1 = pd.read_csv('VIII-16-Entendimiento-de-ingresos-2019.csv', sep=',')
df_1.head()
```

```python
#Eliminar las filas o columnas que tengan solo NA's
df_1 = df_1.dropna(axis=0, how='all')
df_1= df_1.dropna(axis=1, how='all')
df_1
```

```python
#Saber el nombre de las columnas
df_1.columns
```

```python
rename_columns = {
    'MONTH' : 'Month',
    'Fecha de la' : 'Fecha',
    'Ty' : 'Operacion',
    ' Cliente ' : 'Cliente',
    ' Nombre del producto ' : 'Producto',
    ' KG ' : 'KG',
    ' SO Std. Cost ' : 'Costo_Unit_Manufactura',
    ' Precio uni / Venta ' : 'Precio_Unit_Venta',
    ' T.C. ' : 'Tipo_de_Cambio',
    ' Dollars ' : 'Conversion_Ganancia_en_Dolares',
    ' Lbs ' : 'lbs',
    ' Revenue ' : 'Ganancia_MXN',
    ' Foreign Amount ' : 'Ganancia_USD',
    ' CurFrm ' : 'Moneda_de_Venta',
    ' CurTo ' : 'Moneda_MXN',
    'Invoice#' : 'No_Factura',
    ' COST PESOS ' : 'Costo_Tot_Manufactura',
    ' Mes ' : 'Mes'
}
```

```python
#Eliminar y renombrar columnas
df_1 = df_1.drop(columns=[' SalesDiv ',' Std RMCost ',' Std MfgCost ',' Exchange '])
df_1 = df_1.rename(columns = rename_columns)
df_1
```

```python
df_1.dtypes
```

```python
#Cambiar algunas palabras de las columnas y tipos de datos
df_1['Operacion'] = df_1['Operacion'].str.replace('Cancelaci�n', 'Cancelacion')
df_1['No_Factura'] = df_1['No_Factura'].astype(int)
```

```python
df_1['Ganancia_USD'] = df_1['Ganancia_USD'].str.replace(',','')
df_1['Ganancia_USD'] = pd.to_numeric(df_1['Ganancia_USD'], errors = 'coerce', downcast='float')

df_1['Ganancia_MXN'] = df_1['Ganancia_MXN'].str.replace(',','')
df_1['Ganancia_MXN'] = pd.to_numeric(df_1['Ganancia_MXN'], errors = 'coerce', downcast='float')

df_1['Conversion_Ganancia_en_Dolares'] = df_1['Conversion_Ganancia_en_Dolares'].str.replace(',','')
df_1['Conversion_Ganancia_en_Dolares'] = pd.to_numeric(df_1['Conversion_Ganancia_en_Dolares'], errors = 'coerce', downcast='float')

df_1['Costo_Tot_Manufactura'] = df_1['Costo_Tot_Manufactura'].str.replace(',','')
df_1['Costo_Tot_Manufactura'] = pd.to_numeric(df_1['Costo_Tot_Manufactura'], errors = 'coerce', downcast='float')

df_1['lbs'] = df_1['lbs'].str.replace(',','')
df_1['lbs'] = pd.to_numeric(df_1['lbs'], errors = 'coerce', downcast='float')
df_1.dtypes
```

```python
df_1.tail(10)
```

```python
#Volver absolutos los valores de ganancia y kg para evitar errores
df_1['Ganancia_MXN'] = df_1['Ganancia_MXN'].abs()
df_1['Ganancia_USD'] = df_1['Ganancia_USD'].abs() 
df_1['Conversion_Ganancia_en_Dolares'] = df_1['Conversion_Ganancia_en_Dolares'].abs()
df_1['KG'] = df_1['KG'].abs()
df_1
```

```python
def redondeo(value):
    return round(value, 2)
```

```python
#Redondear los floats porque tienen muchos decimales
df_1['Ganancia_MXN'] = df_1['Ganancia_MXN'].map(redondeo)
df_1['Ganancia_USD'] = df_1['Ganancia_USD'].map(redondeo) 
df_1['Conversion_Ganancia_en_Dolares'] = df_1['Conversion_Ganancia_en_Dolares'].map(redondeo)
df_1['Precio_Unit_Venta'] = df_1['Precio_Unit_Venta'].map(redondeo)
df_1['Costo_Unit_Manufactura'] = df_1['Costo_Unit_Manufactura'].map(redondeo)
df_1['Costo_Tot_Manufactura'] = df_1['Costo_Tot_Manufactura'].map(redondeo)
df_1.tail(10)
```

```python
#Buscar mas NA's
df_1.isna().sum(axis=0)
```

```python
#DATAFRAME LIMPIO

#Llenar las NA's con ceros
df_1['KG'] = df_1['KG'].fillna(0)
df_1['Ganancia_USD'] = df_1['Ganancia_USD'].fillna(0)
df_1['Ganancia_MXN'] = df_1['Ganancia_MXN'].fillna(0)
df_1['Precio_Unit_Venta'] = df_1['Precio_Unit_Venta'].fillna(0)
df_1['Conversion_Ganancia_en_Dolares'] = df_1['Conversion_Ganancia_en_Dolares'].fillna(0)
df_1['Costo_Tot_Manufactura'] = df_1['Costo_Tot_Manufactura'].fillna(0)
df_1.isna().sum(axis=0)
```
